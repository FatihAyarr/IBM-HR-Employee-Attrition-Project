# Kategorik verilere one hot encoding uyguluyoruz

```python
df_dum = pd.get_dummies(df_cat)
df_dum.head()
```

# one hot enc uyguladÄ±ÄŸÄ±m df ile numerik verileri birleÅŸtiriyoruz ve features dfimizi elde ediyoruz

```python
X = pd.concat([df_num,df_dum],axis=1)
X.head()
```

# targetÄ±mÄ±z olan attrition kolonunu 0 ve 1 lere Ã§eviriyoruz. 

```python
vals = {"No":0,"Yes":1}
y= y.replace(vals)
print(y) #sadece 0 ve 1 olmalÄ±dÄ±r.
```

# Decision Tree Function

```python
from sklearn import metrics
from sklearn import tree
def predict_dt(X_train,X_test,y_train,y_test):
    clf = tree.DecisionTreeClassifier()
    clf.fit(X_train,y_train)
    
    y_proba = clf.predict_proba(X_test)
    y_pred = clf.predict(X_test)
    
    y_true = y_test
    
    conf_mat = metrics.confusion_matrix(y_true,y_pred)
    x=clf.score(X_test,y_test)
    return (y_true,y_pred,y_proba,x,conf_mat)
```

# Random Forests Function

```python
from sklearn.ensemble import RandomForestClassifier
def predict_rf(X_train,X_test,y_train,y_test):
    clf = RandomForestClassifier()
    clf.fit(X_train,y_train)
    
    y_proba = clf.predict_proba(X_test)
    y_pred = clf.predict(X_test)
    
    y_true = y_test
    
    conf_mat = metrics.confusion_matrix(y_true,y_pred)
    x=clf.score(X_test,y_test)
    return (y_true,y_pred,y_proba,x,conf_mat)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
def predict_lr(X_train,X_test,y_train,y_test):
    clf = LogisticRegression()
    clf.fit(X_train,y_train)
    
    y_proba = clf.predict_proba(X_test)
    y_pred = clf.predict(X_test)
    
    y_true = y_test
    
    conf_mat = metrics.confusion_matrix(y_true,y_pred)
    x=clf.score(X_test,y_test)
    return (y_true,y_pred,y_proba,x,conf_mat)
```

```python
predict_dt(X_train,X_test,y_train,y_test)[3] #accuracy for decision tree
```

```python
predict_rf(X_train,X_test,y_train,y_test)[3] #accuracy for random forest
```
